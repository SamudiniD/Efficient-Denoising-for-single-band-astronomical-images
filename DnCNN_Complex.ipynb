```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# -------------------------------
# 1. Load .npy files (Noisy and Clean images)
# -------------------------------
X = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/noisy.npy")
Y = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/clean.npy")

# Normalize just in case
X = np.clip(X, 0, 1).astype(np.float32)
Y = np.clip(Y, 0, 1).astype(np.float32)

# Add channel dimension if needed
if X.ndim == 3:
    X = X[..., np.newaxis]
if Y.ndim == 3:
    Y = Y[..., np.newaxis]

# -------------------------------
# 2. Train-Test Split
# -------------------------------
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.1, random_state=42)

# -------------------------------
# 3. SSIM + MSE Loss Function
# -------------------------------
def ssim_mse_loss(y_true, y_pred):
    ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_loss = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.5 * mse_loss + 0.5 * ssim_loss

# -------------------------------
# 4. Add Various Types of Noise for Augmentation
# -------------------------------
def add_gaussian_noise(image, std=0.05):
    return np.clip(image + np.random.normal(0, std, image.shape), 0.0, 1.0)

def add_salt_and_pepper(image, salt_prob=0.01, pepper_prob=0.01):
    noisy_image = image.copy()
    total = image.shape[0] * image.shape[1]
    num_salt = int(total * salt_prob)
    num_pepper = int(total * pepper_prob)

    # Salt
    coords = [np.random.randint(0, i, num_salt) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 1.0
    # Pepper
    coords = [np.random.randint(0, i, num_pepper) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 0.0
    return noisy_image

def add_mixed_noise(image):
    return add_gaussian_noise(image, std=0.05) if np.random.rand() < 0.5 else add_salt_and_pepper(image)

# Generate noisy input-target pairs
X_augmented = np.array([add_mixed_noise(img) for img in X])
Y_augmented = np.array([add_mixed_noise(img) for img in Y])

# -------------------------------
# 5. DnCNN Model with Dropout and Regularization
# -------------------------------
def build_dncnn(input_shape=(240, 240, 1), num_filters=64, depth=19, dropout_rate=0.3):
    """
    Build the DnCNN model with increased depth, regularization, and dropout.
    """
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(num_filters, 3, padding="same", activation="relu", kernel_regularizer=regularizers.l2(0.01))(inp)

    # Increased depth by adding more convolution layers
    for _ in range(depth - 2):
        x = layers.Conv2D(num_filters, 3, padding="same", kernel_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)  # Dropout after each block for regularization

    # Final convolution layer to predict the noise (output)
    x = layers.Conv2D(1, 3, padding="same")(x)
    out = layers.Subtract()([inp, x])  # Subtract residual noise
    return models.Model(inp, out)

# -------------------------------
# 6. Compile and Train the Model
# -------------------------------
model = build_dncnn()
model.compile(optimizer=optimizers.Adam(1e-5), loss=ssim_mse_loss)  # Initial learning rate

# Set up callbacks
callbacks = [
    ModelCheckpoint('/content/drive/MyDrive/PS1_Models/dncnn_best.h5', save_best_only=True),
    EarlyStopping(patience=20, restore_best_weights=True),  # Increased patience
    ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6)  # Reduce learning rate if the validation loss stops improving
]

# Train the model
history = model.fit(
    X_augmented, Y_augmented,
    validation_data=(X_val, Y_val),
    epochs=100,
    batch_size=8,
    callbacks=callbacks,
    shuffle=True
)

# -------------------------------
# 7. Save Final Model
# -------------------------------
model.save('/content/drive/MyDrive/PS1_Models/dncnn_complex.h5')
print("✅ Model trained and saved!")

# -------------------------------
# 8. Plot Training History
# -------------------------------
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("DnCNN Training Loss (Augmented Noise)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------
# 9. Model Evaluation: PSNR and SSIM
# -------------------------------
import numpy as np
import tensorflow as tf
from astropy.io import fits
from tensorflow.keras.models import load_model
from skimage.metrics import peak_signal_noise_ratio as psnr_metric
from skimage.metrics import structural_similarity as ssim_metric

# Load the trained model
model_path = "/content/drive/MyDrive/PS1_Models/dncnn_complex.h5"
model = load_model(model_path, custom_objects={'ssim_mse_loss': ssim_mse_loss})
print("✅ Model loaded!")

# -----------------------------
# 3. Load FITS Image Pair
# -----------------------------
noisy_path = "/content/drive/MyDrive/PS1_Dataset/renamed/noisy/galaxy_004_noisy.fits"
clean_path = "/content/drive/MyDrive/PS1_Dataset/renamed/clean/galaxy_004_clean.fits"

def load_fits_normalized(path):
    with fits.open(path) as hdul:
        img = hdul[0].data.astype(np.float32)
        img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0)
        p1, p99 = np.percentile(img, (1, 99))
        img = np.clip(img, p1, p99)
        img = (img - p1) / (p99 - p1 + 1e-8)
        img = np.log1p(img) / np.log1p(1.0)
        return img

noisy_img = load_fits_normalized(noisy_path)
clean_img = load_fits_normalized(clean_path)

# Resize to 240x240 for model input
noisy_resized = tf.image.resize(noisy_img[..., np.newaxis], (240, 240)).numpy().squeeze()
clean_resized = tf.image.resize(clean_img[..., np.newaxis], (240, 240)).numpy().squeeze()

# -----------------------------
# 4. Predict denoised output
# -----------------------------
input_tensor = noisy_resized[np.newaxis, ..., np.newaxis]
denoised = model.predict(input_tensor, verbose=0)[0, ..., 0]

# -----------------------------
# 5. Evaluation Metrics
# -----------------------------
psnr = psnr_metric(clean_resized, denoised, data_range=1.0)
ssim = ssim_metric(clean_resized, denoised, data_range=1.0)

print("\n📊 Evaluation Metrics:")
print(f"PSNR: {psnr:.2f} dB")
print(f"SSIM: {ssim:.4f}")

# -----------------------------
# 6. Visualization
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, img, title in zip(
    axes,
    [noisy_resized, denoised, clean_resized],
    ["Noisy FITS", f"Denoised Output\nPSNR: {psnr:.2f} dB | SSIM: {ssim:.4f}", "Clean FITS (Reference)"]
):
    ax.imshow(img, cmap="gray", vmin=np.percentile(img, 1), vmax=np.percentile(img, 99))
    ax.set_title(title)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

# DnCNN-Based Galaxy Image Denoising Pipeline

 This cell implements an end-to-end workflow for training, evaluating, and
 visualizing a DnCNN denoising model on astronomical images:

# 1. Dataset Loading & Augmentation
    -Loads noisy/clean .npy arrays from SDSS training data
    - Clips to [0,1], adds channel if needed
    - Applies mixed noise augmentation (Gaussian + Salt & Pepper)

# 2. Model Definition & Loss
    - Builds a 19-layer DnCNN with L2 regularization and dropout
    - Uses a combined SSIM + MSE loss for balanced noise removal + structure preservation

# 3. Training with Timing
    - Splits augmented data into train/validation (90/10)
    - Trains for up to 100 epochs with callbacks:
        • ModelCheckpoint (save best by val_loss)
        • EarlyStopping (patience=20)
        • ReduceLROnPlateau (patience=5, factor=0.5)
    - Records total training time and prints it in minutes
    - Saves final model as 'dncnn_complex.h5'

# 4. Training Curve Plot
    - Plots train vs. validation loss over epochs

# 5. Model Reload
    - Loads the best-saved model with custom loss

# 6. FITS Test-Set Evaluation with Timing
    - Reads all FITS files in PS1 noisy/clean test folders
    - Normalizes by 1st/99th percentiles + log1p scaling
    - Runs inference on each image, times total and per-image inference
    - Computes PSNR and SSIM per image#    - Prints:
       • Number of test images
       • Total & average inference time
       • Mean ± std and 95% CI for PSNR (dB) and SSIM

# 7. PSNR & SSIM Distribution Plots
    - Histograms of PSNR and SSIM with red dashed mean lines

# 8. Mean ±1σ Bar Chart
    - Bar chart of PSNR and SSIM means with error bars = ±1 standard deviation

# 9. Galaxy_004 Example
    - Loads the specific noisy/clean FITS pair for 'galaxy_004'
    - Times single-image inference
    - Computes & prints its PSNR/SSIM
    - Displays noisy ∣ denoised ∣ clean side-by-side

```python
import glob
import time

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from scipy import stats
from astropy.io import fits
from skimage.metrics import (
    peak_signal_noise_ratio as psnr_metric,
    structural_similarity as ssim_metric
)
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import (
    ModelCheckpoint,
    EarlyStopping,
    ReduceLROnPlateau
)

# ------------------------------- 1. Dataset Loading & Augmentation -------------------------------
def load_dataset():
    X = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/noisy.npy")
    Y = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/clean.npy")
    X = np.clip(X, 0, 1).astype(np.float32)
    Y = np.clip(Y, 0, 1).astype(np.float32)
    if X.ndim == 3: X = X[..., np.newaxis]
    if Y.ndim == 3: Y = Y[..., np.newaxis]
    return X, Y

def add_gaussian_noise(image, std=0.05):
    return np.clip(image + np.random.normal(0, std, image.shape), 0, 1)

def add_salt_and_pepper(image, salt_prob=0.01, pepper_prob=0.01):
    noisy = image.copy()
    total = image.shape[0] * image.shape[1]
    # salt
    coords = [np.random.randint(0, i, int(total*salt_prob)) for i in image.shape[:2]]
    noisy[coords[0], coords[1], 0] = 1.0
    # pepper
    coords = [np.random.randint(0, i, int(total*pepper_prob)) for i in image.shape[:2]]
    noisy[coords[0], coords[1], 0] = 0.0
    return noisy

def add_mixed_noise(image):
    return add_gaussian_noise(image) if np.random.rand()<0.5 else add_salt_and_pepper(image)

def augment_data(X_clean):
    X_noisy = np.array([add_mixed_noise(img) for img in X_clean])
    Y_clean = X_clean.copy()
    return X_noisy, Y_clean

# ------------------------------- 2. Loss & Model Builder -------------------------------
def ssim_mse_loss(y_true, y_pred):
    ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_loss  = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.5*mse_loss + 0.5*ssim_loss

def build_dncnn(input_shape=(240,240,1), num_filters=64, depth=19, dropout_rate=0.3):
    inp = layers.Input(shape=input_shape)
    x   = layers.Conv2D(num_filters,3,padding="same",activation="relu",
                        kernel_regularizer=regularizers.l2(0.01))(inp)
    for _ in range(depth-2):
        x = layers.Conv2D(num_filters,3,padding="same",
                          kernel_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)
    x   = layers.Conv2D(1,3,padding="same")(x)
    out = layers.Subtract()([inp, x])
    return models.Model(inp, out)

# ------------------------------- 3. Prepare & Train -------------------------------
# Load & augment
X, Y            = load_dataset()
X_aug, Y_aug    = augment_data(Y)  # feed in clean Y to get X_noisy, Y_clean
X_tr, X_val, Y_tr, Y_val = train_test_split(
    X_aug, Y_aug, test_size=0.1, random_state=42
)

model = build_dncnn()
model.compile(optimizer=optimizers.Adam(1e-5), loss=ssim_mse_loss)

callbacks = [
    ModelCheckpoint(
        '/content/drive/MyDrive/PS1_Models/dncnn_best.h5',
        save_best_only=True, monitor='val_loss'
    ),
    EarlyStopping(patience=20, restore_best_weights=True),
    ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6)
]

print("⏱️ Training started...")
t_start = time.time()

history = model.fit(
    X_tr, Y_tr,
    validation_data=(X_val, Y_val),
    epochs=100, batch_size=8,
    callbacks=callbacks, shuffle=True,
    verbose=1
)

t_end = time.time()
train_time = t_end - t_start
print(f"✅ Training finished in {train_time/60:.2f} minutes")

# Save final
model.save('/content/drive/MyDrive/PS1_Models/dncnn_complex.h5')

# Plot loss curves
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title("DnCNN Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------- 4. Reload Best Model -------------------------------
model = tf.keras.models.load_model(
    '/content/drive/MyDrive/PS1_Models/dncnn_complex.h5',
    custom_objects={'ssim_mse_loss': ssim_mse_loss}
)

# ------------------------------- 5. FITS Evaluation -------------------------------
def load_fits_normalized(path):
    with fits.open(path) as hdul:
        img = hdul[0].data.astype(np.float32)
    img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0)
    p1, p99 = np.percentile(img, (1,99))
    img = np.clip(img, p1, p99)
    img = (img - p1)/(p99 - p1 + 1e-8)
    return np.log1p(img)/np.log1p(1.0)

def preprocess(img, size=(240,240)):
    return tf.image.resize(img[...,np.newaxis], size).numpy()

# collect test file lists
noisy_paths = sorted(glob.glob(
    '/content/drive/MyDrive/PS1_Dataset/renamed/noisy/*.fits'
))
clean_paths = sorted(glob.glob(
    '/content/drive/MyDrive/PS1_Dataset/renamed/clean/*.fits'
))
assert len(noisy_paths)==len(clean_paths)

psnr_list, ssim_list = [], []
t0 = time.time()
for n_p, c_p in zip(noisy_paths, clean_paths):
    noisy = load_fits_normalized(n_p)
    clean = load_fits_normalized(c_p)
    inp   = preprocess(noisy)[np.newaxis,...]
    den   = model.predict(inp, verbose=0)[0,...,0]
    psnr_list.append(psnr_metric(clean, den, data_range=1.0))
    ssim_list.append(ssim_metric(clean, den, data_range=1.0))
t1 = time.time()

# stats
psnr_arr = np.array(psnr_list)
ssim_arr = np.array(ssim_list)
n_imgs   = len(psnr_arr)
inf_time_total = t1-t0
inf_time_avg   = inf_time_total/n_imgs

psnr_mu, psnr_sd = psnr_arr.mean(), psnr_arr.std()
ssim_mu, ssim_sd = ssim_arr.mean(), ssim_arr.std()
ci_psnr = stats.norm.interval(0.95, psnr_mu, psnr_sd/np.sqrt(n_imgs))
ci_ssim = stats.norm.interval(0.95, ssim_mu, ssim_sd/np.sqrt(n_imgs))

print(f"▶️ Inference on {n_imgs} images")
print(f"   • Total inference time: {inf_time_total:.2f}s  (avg {inf_time_avg:.4f}s/img)")
print(f"   • PSNR  mean={psnr_mu:.2f} dB ±{psnr_sd:.2f}, 95% CI={ci_psnr}")
print(f"   • SSIM  mean={ssim_mu:.4f} ±{ssim_sd:.4f}, 95% CI={ci_ssim}")

# ------------------------------- 6. Visualize One Example -------------------------------
# use exactly your provided paths:
noisy_example = load_fits_normalized(
    "/content/drive/MyDrive/PS1_Dataset/renamed/noisy/galaxy_004_noisy.fits"
)
clean_example = load_fits_normalized(
    "/content/drive/MyDrive/PS1_Dataset/renamed/clean/galaxy_004_clean.fits"
)

inp_ex = preprocess(noisy_example)[np.newaxis,...]
den_ex = model.predict(inp_ex, verbose=0)[0,...,0]

psnr_ex = psnr_metric(clean_example, den_ex, data_range=1.0)
ssim_ex = ssim_metric(clean_example, den_ex, data_range=1.0)

fig, axs = plt.subplots(1,3,figsize=(12,4))
for ax, img, title in zip(
    axs,
    [noisy_example, den_ex, clean_example],
    [
      "Noisy FITS",
      f"Denoised\nPSNR={psnr_ex:.2f} dB\nSSIM={ssim_ex:.4f}",
      "Clean FITS"
    ]
):
    ax.imshow(img, cmap='gray',
              vmin=np.percentile(img,1),
              vmax=np.percentile(img,99))
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import glob
import time

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from scipy import stats
from astropy.io import fits
from skimage.metrics import (
    peak_signal_noise_ratio as psnr_metric,
    structural_similarity as ssim_metric
)
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import (
    ModelCheckpoint,
    EarlyStopping,
    ReduceLROnPlateau
)

# ------------------------------- 1. Dataset Loading & Augmentation -------------------------------
def load_dataset():
    X = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/noisy.npy")
    Y = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/clean.npy")
    X = np.clip(X, 0, 1).astype(np.float32)
    Y = np.clip(Y, 0, 1).astype(np.float32)
    if X.ndim == 3: X = X[..., np.newaxis]
    if Y.ndim == 3: Y = Y[..., np.newaxis]
    return X, Y

def add_gaussian_noise(image, std=0.05):
    return np.clip(image + np.random.normal(0, std, image.shape), 0, 1)

def add_salt_and_pepper(image, salt_prob=0.01, pepper_prob=0.01):
    noisy = image.copy()
    total = image.shape[0] * image.shape[1]
    coords = [np.random.randint(0, i, int(total*salt_prob)) for i in image.shape[:2]]
    noisy[coords[0], coords[1], 0] = 1.0
    coords = [np.random.randint(0, i, int(total*pepper_prob)) for i in image.shape[:2]]
    noisy[coords[0], coords[1], 0] = 0.0
    return noisy

def add_mixed_noise(image):
    return add_gaussian_noise(image) if np.random.rand()<0.5 else add_salt_and_pepper(image)

def augment_data(X_clean):
    X_noisy = np.array([add_mixed_noise(img) for img in X_clean])
    Y_clean = X_clean.copy()
    return X_noisy, Y_clean

# ------------------------------- 2. Loss & Model Builder -------------------------------
def ssim_mse_loss(y_true, y_pred):
    ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_loss  = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.5*mse_loss + 0.5*ssim_loss

def build_dncnn(input_shape=(240,240,1), num_filters=64, depth=19, dropout_rate=0.3):
    inp = layers.Input(shape=input_shape)
    x   = layers.Conv2D(num_filters,3,padding="same",activation="relu",
                        kernel_regularizer=regularizers.l2(0.01))(inp)
    for _ in range(depth-2):
        x = layers.Conv2D(num_filters,3,padding="same",
                          kernel_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)
    x   = layers.Conv2D(1,3,padding="same")(x)
    out = layers.Subtract()([inp, x])
    return models.Model(inp, out)

# ------------------------------- 3. Prepare & Train -------------------------------
X, Y            = load_dataset()
X_aug, Y_aug    = augment_data(Y)
X_tr, X_val, Y_tr, Y_val = train_test_split(X_aug, Y_aug, test_size=0.1, random_state=42)

model = build_dncnn()
model.compile(optimizer=optimizers.Adam(1e-5), loss=ssim_mse_loss)

callbacks = [
    ModelCheckpoint('/content/drive/MyDrive/PS1_Models/dncnn_best.h5', save_best_only=True),
    EarlyStopping(patience=20, restore_best_weights=True),
    ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6)
]

print("⏱️ Training started...")
t_train_start = time.time()
history = model.fit(
    X_tr, Y_tr,
    validation_data=(X_val, Y_val),
    epochs=100, batch_size=8,
    callbacks=callbacks, shuffle=True, verbose=1
)
t_train_end = time.time()
train_time = t_train_end - t_train_start
print(f"✅ Training finished in {train_time/60:.2f} minutes")

model.save('/content/drive/MyDrive/PS1_Models/dncnn_complex.h5')

# ------------------------------- 4. Plot Training Curve -------------------------------
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("DnCNN Training & Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------- 5. Reload Best Model -------------------------------
model = tf.keras.models.load_model(
    '/content/drive/MyDrive/PS1_Models/dncnn_complex.h5',
    custom_objects={'ssim_mse_loss': ssim_mse_loss}
)

# ------------------------------- 6. FITS Evaluation -------------------------------
def load_fits_normalized(path):
    with fits.open(path) as hdul:
        img = hdul[0].data.astype(np.float32)
    img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0)
    p1, p99 = np.percentile(img, (1,99))
    img = np.clip(img, p1, p99)
    img = (img - p1)/(p99 - p1 + 1e-8)
    return np.log1p(img)/np.log1p(1.0)

def preprocess(img, size=(240,240)):
    return tf.image.resize(img[...,np.newaxis], size).numpy()

noisy_paths = sorted(glob.glob('/content/drive/MyDrive/PS1_Dataset/renamed/noisy/*.fits'))
clean_paths = sorted(glob.glob('/content/drive/MyDrive/PS1_Dataset/renamed/clean/*.fits'))
assert len(noisy_paths)==len(clean_paths)

psnr_list, ssim_list = [], []
t_inf_start = time.time()
for npath, cpath in zip(noisy_paths, clean_paths):
    nimg = load_fits_normalized(npath)
    cimg = load_fits_normalized(cpath)
    den  = model.predict(preprocess(nimg)[np.newaxis,...], verbose=0)[0,...,0]
    psnr_list.append(psnr_metric(cimg, den, data_range=1.0))
    ssim_list.append(ssim_metric(cimg, den, data_range=1.0))
t_inf_end = time.time()
inf_time_total = t_inf_end - t_inf_start

psnr_arr = np.array(psnr_list)
ssim_arr = np.array(ssim_list)
n_imgs   = len(psnr_arr)
inf_time_avg = inf_time_total / n_imgs

psnr_mu, psnr_sd = psnr_arr.mean(), psnr_arr.std()
ssim_mu, ssim_sd = ssim_arr.mean(), ssim_arr.std()
ci_psnr = stats.norm.interval(0.95, loc=psnr_mu, scale=psnr_sd/np.sqrt(n_imgs))
ci_ssim = stats.norm.interval(0.95, loc=ssim_mu, scale=ssim_sd/np.sqrt(n_imgs))

print(f"▶️ Inference on {n_imgs} images")
print(f"   • Total inference time: {inf_time_total:.2f}s  (avg {inf_time_avg:.4f}s/img)")
print(f"   • PSNR mean={psnr_mu:.2f}±{psnr_sd:.2f}, 95% CI={ci_psnr}")
print(f"   • SSIM mean={ssim_mu:.4f}±{ssim_sd:.4f}, 95% CI={ci_ssim}")

# ------------------------------- 7. PSNR & SSIM Distribution Plots -------------------------------
plt.figure(figsize=(6,4))
plt.hist(psnr_arr, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(psnr_mu, color='red', linestyle='--', label=f"Mean = {psnr_mu:.2f} dB")
plt.title('PSNR Distribution')
plt.xlabel('PSNR (dB)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.hist(ssim_arr, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(ssim_mu, color='red', linestyle='--', label=f"Mean = {ssim_mu:.3f}")
plt.title('SSIM Distribution')
plt.xlabel('SSIM')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------- 8. Mean ±1σ Bar Chart -------------------------------
means = [psnr_mu, ssim_mu]
stds  = [psnr_sd, ssim_sd]
labels = ['PSNR (dB)', 'SSIM']

plt.figure(figsize=(5,4))
bars = plt.bar(labels, means, yerr=stds, capsize=8, edgecolor='black', alpha=0.8)
for bar, m in zip(bars, means):
    plt.text(bar.get_x() + bar.get_width()/2, m + 0.02*max(means), f"{m:.2f}",
             ha='center', va='bottom')
plt.title('Test-Set Mean ± 1σ')
plt.ylabel('Value')
plt.tight_layout()
plt.show()

# ------------------------------- 9. Visualize Galaxy_004 Example -------------------------------
noisy_004 = load_fits_normalized("/content/drive/MyDrive/PS1_Dataset/renamed/noisy/galaxy_004_noisy.fits")
clean_004 = load_fits_normalized("/content/drive/MyDrive/PS1_Dataset/renamed/clean/galaxy_004_clean.fits")
inp_004 = preprocess(noisy_004)[np.newaxis,...]

t_single_start = time.time()
den_004 = model.predict(inp_004, verbose=0)[0,...,0]
t_single_end = time.time()
single_time = t_single_end - t_single_start

psnr_004 = psnr_metric(clean_004, den_004, data_range=1.0)
ssim_004 = ssim_metric(clean_004, den_004, data_range=1.0)

print(f"\nGalaxy_004 Inference time: {single_time:.4f}s")
print(f"PSNR_004 = {psnr_004:.2f} dB, SSIM_004 = {ssim_004:.4f}\n")

fig, axs = plt.subplots(1,3,figsize=(12,4))
for ax, img, title in zip(
    axs,
    [noisy_004, den_004, clean_004],
    ["Noisy FITS", f"Denoised\nPSNR={psnr_004:.2f} dB\nSSIM={ssim_004:.4f}", "Clean FITS"]
):
    ax.imshow(img, cmap='gray',
              vmin=np.percentile(img,1), vmax=np.percentile(img,99))
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# -------------------------------
# 1. Load .npy files (Noisy and Clean images)
# -------------------------------
X = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/noisy.npy")
Y = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/clean.npy")

# Normalize just in case
X = np.clip(X, 0, 1).astype(np.float32)
Y = np.clip(Y, 0, 1).astype(np.float32)

# Add channel dimension if needed
if X.ndim == 3:
    X = X[..., np.newaxis]
if Y.ndim == 3:
    Y = Y[..., np.newaxis]

# -------------------------------
# 2. Train-Test Split
# -------------------------------
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.1, random_state=42)

# -------------------------------
# 3. SSIM + MSE Loss Function
# -------------------------------
def ssim_mse_loss(y_true, y_pred):
    ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_loss = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.5 * mse_loss + 0.5 * ssim_loss

# -------------------------------
# 4. Add Various Types of Noise for Augmentation
# -------------------------------
def add_gaussian_noise(image, std=0.05):
    return np.clip(image + np.random.normal(0, std, image.shape), 0.0, 1.0)

def add_salt_and_pepper(image, salt_prob=0.01, pepper_prob=0.01):
    noisy_image = image.copy()
    total = image.shape[0] * image.shape[1]
    num_salt = int(total * salt_prob)
    num_pepper = int(total * pepper_prob)

    # Salt
    coords = [np.random.randint(0, i, num_salt) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 1.0
    # Pepper
    coords = [np.random.randint(0, i, num_pepper) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 0.0
    return noisy_image

def add_poisson_noise(image):
    return np.clip(np.random.poisson(image * 255) / 255.0, 0, 1)

def add_mixed_noise(image):
    if np.random.rand() < 0.5:
        return add_gaussian_noise(image, std=np.random.uniform(0.01, 0.1))  # Random std for Gaussian
    elif np.random.rand() < 0.5:
        return add_salt_and_pepper(image, salt_prob=np.random.uniform(0.005, 0.02), pepper_prob=np.random.uniform(0.005, 0.02))
    else:
        return add_poisson_noise(image)

# Generate noisy input-target pairs with diverse noise
X_augmented = np.array([add_mixed_noise(img) for img in X])
Y_augmented = np.array([add_mixed_noise(img) for img in Y])

# -------------------------------
# 5. DnCNN Model with Increased Depth and Filters
# -------------------------------
def build_dncnn(input_shape=(240, 240, 1), num_filters=128, depth=21, dropout_rate=0.3):
    """
    Build the DnCNN model with increased depth, filters, and regularization.
    """
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(num_filters, 3, padding="same", activation="relu", kernel_regularizer=regularizers.l2(0.01))(inp)

    # Increased depth by adding more convolution layers
    for _ in range(depth - 2):
        x = layers.Conv2D(num_filters, 3, padding="same", kernel_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)  # Dropout after each block for regularization

    # Final convolution layer to predict the noise (output)
    x = layers.Conv2D(1, 3, padding="same")(x)
    out = layers.Subtract()([inp, x])  # Subtract residual noise
    return models.Model(inp, out)

# -------------------------------
# 6. Compile and Train the Model
# -------------------------------
model = build_dncnn()
model.compile(optimizer=optimizers.Adam(learning_rate=1e-4), loss=ssim_mse_loss)  # Increased learning rate

# Set up callbacks
callbacks = [
    ModelCheckpoint('/content/drive/MyDrive/PS1_Models/dncnn_best.h5', save_best_only=True),
    EarlyStopping(patience=20, restore_best_weights=True),  # Increased patience
    ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6)  # Reduce learning rate if the validation loss stops improving
]

# Train the model
history = model.fit(
    X_augmented, Y_augmented,
    validation_data=(X_val, Y_val),
    epochs=100,
    batch_size=8,
    callbacks=callbacks,
    shuffle=True
)

# -------------------------------
# 7. Save Final Model
# -------------------------------
model.save('/content/drive/MyDrive/PS1_Models/dncnn_t1.h5')
print("✅ Model trained and saved!")

# -------------------------------
# 8. Plot Training History
# -------------------------------
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("DnCNN Training Loss (Augmented Noise)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------
# 9. Model Evaluation: PSNR and SSIM
# -------------------------------
import numpy as np
import tensorflow as tf
from astropy.io import fits
from tensorflow.keras.models import load_model
from skimage.metrics import peak_signal_noise_ratio as psnr_metric
from skimage.metrics import structural_similarity as ssim_metric

# Load the trained model
model_path = "/content/drive/MyDrive/PS1_Models/dncnn_t1.h5"
model = load_model(model_path, custom_objects={'ssim_mse_loss': ssim_mse_loss})
print("✅ Model loaded!")

# -----------------------------
# 3. Load FITS Image Pair
# -----------------------------
noisy_path = "/content/drive/MyDrive/PS1_Dataset/renamed/noisy/galaxy_004_noisy.fits"
clean_path = "/content/drive/MyDrive/PS1_Dataset/renamed/clean/galaxy_004_clean.fits"

def load_fits_normalized(path):
    with fits.open(path) as hdul:
        img = hdul[0].data.astype(np.float32)
        img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0)
        p1, p99 = np.percentile(img, (1, 99))
        img = np.clip(img, p1, p99)
        img = (img - p1) / (p99 - p1 + 1e-8)
        img = np.log1p(img) / np.log1p(1.0)
        return img

noisy_img = load_fits_normalized(noisy_path)
clean_img = load_fits_normalized(clean_path)

# Resize to 240x240 for model input
noisy_resized = tf.image.resize(noisy_img[..., np.newaxis], (240, 240)).numpy().squeeze()
clean_resized = tf.image.resize(clean_img[..., np.newaxis], (240, 240)).numpy().squeeze()

# -----------------------------
# 4. Predict denoised output
# -----------------------------
input_tensor = noisy_resized[np.newaxis, ..., np.newaxis]
denoised = model.predict(input_tensor, verbose=0)[0, ..., 0]

# -----------------------------
# 5. Evaluation Metrics
# -----------------------------
psnr = psnr_metric(clean_resized, denoised, data_range=1.0)
ssim = ssim_metric(clean_resized, denoised, data_range=1.0)

print("\n📊 Evaluation Metrics:")
print(f"PSNR: {psnr:.2f} dB")
print(f"SSIM: {ssim:.4f}")

# -----------------------------
# 6. Visualization
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, img, title in zip(
    axes,
    [noisy_resized, denoised, clean_resized],
    ["Noisy FITS", f"Denoised Output\nPSNR: {psnr:.2f} dB | SSIM: {ssim:.4f}", "Clean FITS (Reference)"]
):
    ax.imshow(img, cmap="gray", vmin=np.percentile(img, 1), vmax=np.percentile(img, 99))
    ax.set_title(title)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from astropy.io import fits
from skimage.metrics import peak_signal_noise_ratio as psnr_metric
from skimage.metrics import structural_similarity as ssim_metric

# -------------------------------
# 1. Load .npy files (Noisy and Clean images)
# -------------------------------
X = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/noisy.npy")
Y = np.load("/content/drive/MyDrive/SDSS_augmented_npy/train/clean.npy")

# Normalize to [0, 1] using Min-Max Normalization
X = np.clip(X, 0, 1).astype(np.float32)
Y = np.clip(Y, 0, 1).astype(np.float32)

# Add channel dimension if needed
if X.ndim == 3:
    X = X[..., np.newaxis]
if Y.ndim == 3:
    Y = Y[..., np.newaxis]

# -------------------------------
# 2. Train-Test Split
# -------------------------------
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.1, random_state=42)

# -------------------------------
# 3. SSIM + MSE Loss Function
# -------------------------------
def ssim_mse_loss(y_true, y_pred):
    ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_loss = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.5 * mse_loss + 0.5 * ssim_loss

# -------------------------------
# 4. Add Various Types of Noise for Augmentation
# -------------------------------
def add_gaussian_noise(image, std=0.05):
    return np.clip(image + np.random.normal(0, std, image.shape), 0.0, 1.0)

def add_salt_and_pepper(image, salt_prob=0.01, pepper_prob=0.01):
    noisy_image = image.copy()
    total = image.shape[0] * image.shape[1]
    num_salt = int(total * salt_prob)
    num_pepper = int(total * pepper_prob)

    # Salt
    coords = [np.random.randint(0, i, num_salt) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 1.0
    # Pepper
    coords = [np.random.randint(0, i, num_pepper) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1], 0] = 0.0
    return noisy_image

def add_poisson_noise(image):
    return np.clip(np.random.poisson(image * 255) / 255.0, 0, 1)

def add_mixed_noise(image):
    if np.random.rand() < 0.5:
        return add_gaussian_noise(image, std=np.random.uniform(0.01, 0.1))  # Random std for Gaussian
    elif np.random.rand() < 0.5:
        return add_salt_and_pepper(image, salt_prob=np.random.uniform(0.005, 0.02), pepper_prob=np.random.uniform(0.005, 0.02))
    else:
        return add_poisson_noise(image)

# Generate noisy input-target pairs with diverse noise
X_augmented = np.array([add_mixed_noise(img) for img in X])
Y_augmented = np.array([add_mixed_noise(img) for img in Y])

# -------------------------------
# 5. DnCNN Model with Increased Depth and Filters
# -------------------------------
def build_dncnn(input_shape=(240, 240, 1), num_filters=128, depth=21, dropout_rate=0.3):
    """
    Build the DnCNN model with increased depth, filters, and regularization.
    """
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(num_filters, 3, padding="same", activation="relu", kernel_regularizer=regularizers.l2(0.01))(inp)

    # Increased depth by adding more convolution layers
    for _ in range(depth - 2):
        x = layers.Conv2D(num_filters, 3, padding="same", kernel_regularizer=regularizers.l2(0.01))(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Dropout(dropout_rate)(x)  # Dropout after each block for regularization

    # Final convolution layer to predict the noise (output)
    x = layers.Conv2D(1, 3, padding="same")(x)
    out = layers.Subtract()([inp, x])  # Subtract residual noise
    return models.Model(inp, out)

# -------------------------------
# 6. Compile and Train the Model
# -------------------------------
model = build_dncnn()
model.compile(optimizer=optimizers.Adam(learning_rate=1e-4), loss=ssim_mse_loss)  # Increased learning rate

# Set up callbacks
callbacks = [
    ModelCheckpoint('/content/drive/MyDrive/PS1_Models/dncnn_best.h5', save_best_only=True),
    EarlyStopping(patience=30, restore_best_weights=True),  # Increased patience for better convergence
    ReduceLROnPlateau(patience=5, factor=0.5, min_lr=1e-6)  # Reduce learning rate if the validation loss stops improving
]

# Train the model for more epochs (200 or 300)
history = model.fit(
    X_augmented, Y_augmented,
    validation_data=(X_val, Y_val),
    epochs=150,
    batch_size=16,  # Increased batch size
    callbacks=callbacks,
    shuffle=True
)

# -------------------------------
# 7. Save Final Model
# -------------------------------
model.save('/content/drive/MyDrive/PS1_Models/dncnn_200epochs.h5')
print("✅ Model trained and saved!")

# -------------------------------
# 8. Plot Training History
# -------------------------------
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("DnCNN Training Loss (Augmented Noise)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------
# 9. Model Evaluation: PSNR and SSIM
# -------------------------------
# Load the trained model
model_path = "/content/drive/MyDrive/PS1_Models/dncnn_200epochs.h5"
model = load_model(model_path, custom_objects={'ssim_mse_loss': ssim_mse_loss})
print("✅ Model loaded!")

# -----------------------------
# 3. Load FITS Image Pair
# -----------------------------
noisy_path = "/content/drive/MyDrive/PS1_Dataset/renamed/noisy/galaxy_004_noisy.fits"
clean_path = "/content/drive/MyDrive/PS1_Dataset/renamed/clean/galaxy_004_clean.fits"

def load_fits_normalized(path):
    with fits.open(path) as hdul:
        img = hdul[0].data.astype(np.float32)
        img = np.nan_to_num(img, nan=0.0, posinf=0.0, neginf=0.0)

        # Min-Max Normalization
        pmin, pmax = np.percentile(img, (1, 99))  # Use 1st and 99th percentile for better contrast
        img = np.clip(img, pmin, pmax)  # Clip to remove outliers
        img = (img - pmin) / (pmax - pmin + 1e-8)  # Normalize to [0, 1]

        return img

noisy_img = load_fits_normalized(noisy_path)
clean_img = load_fits_normalized(clean_path)

# Resize to 240x240 for model input
noisy_resized = tf.image.resize(noisy_img[..., np.newaxis], (240, 240)).numpy().squeeze()
clean_resized = tf.image.resize(clean_img[..., np.newaxis], (240, 240)).numpy().squeeze()

# -----------------------------
# 4. Predict denoised output
# -----------------------------
input_tensor = noisy_resized[np.newaxis, ..., np.newaxis]
denoised = model.predict(input_tensor, verbose=0)[0, ..., 0]

# Post-processing: Ensure denoised image is within [0, 1]
denoised = np.clip(denoised, 0.0, 1.0)  # Clip values to the range [0, 1]

# -----------------------------
# 5. Evaluation Metrics
# -----------------------------
psnr = psnr_metric(clean_resized, denoised, data_range=1.0)
ssim = ssim_metric(clean_resized, denoised, data_range=1.0)

print("\n📊 Evaluation Metrics:")
print(f"PSNR: {psnr:.2f} dB")
print(f"SSIM: {ssim:.4f}")

# -----------------------------
# 6. Visualization
# -----------------------------
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, img, title in zip(
    axes,
    [noisy_resized, denoised, clean_resized],
    ["Noisy FITS", f"Denoised Output\nPSNR: {psnr:.2f} dB | SSIM: {ssim:.4f}", "Clean FITS (Reference)"]
):
    ax.imshow(img, cmap="gray", vmin=np.percentile(img, 1), vmax=np.percentile(img, 99))
    ax.set_title(title)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
